```python
import random

# Add a ":" after the inout statement
age1=int(input("first player please input your age: "))
age2=int(input("second player please input your age: "))
if age1<age2:
    print("first player begins")
else:
    print("second player begins")

# Define the result of rolling 5 dice at once
def roll_dice (num_dice):
    dice_result=[]
    for i in range (0,num_dice):
        d=random.randint(1,6)
        dice_result.append(d)
    return dice_result

# Player 1 rolls the dice
remain_number=5
# Final score list
l_m=[]
# Temporary result list 
l6=[]
while remain_number>0:
    result=roll_dice(remain_number)
    for i in range(len(result)):
        if result[i]!=6:
            l6.append(result[i])
    print(l6)
    m=max(l6)      
    l_m.append(m)
    remain_number-=1
print(l_m)
total_first_player = sum(l_m)
print(total_first_player)

# Player 2 rolls the dice
print("second player")
remain_number=5
# Final score list
l_m=[]
# Temporary result list
l6=[]
while remain_number>0:
    result=roll_dice(remain_number)
    for i in range(len(result)):
        if result[i]!=6:
            l6.append(result[i])     
    print(l6)
    m=max(l6)      
    l_m.append(m)
    remain_number-=1
print(l_m)
total_second_player = sum(l_m)
print(total_second_player) 

# Decide who wins based on the Final score list of each player
if total_first_player > total_second_player:
    print("first player wins")
elif total_first_player < total_second_player:
    print("second player wins")
else:
    print("there is no winner")
```
